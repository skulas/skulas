### Parse to json a XDR log

```python
output_1 = subprocess.check_output(['xdrparser', 'XDRs/bucket-002ac59efd06e1568fff3b38cde1d369eb9e17debd1dd1ce77520cb67d515200.xdr'])
output_2 = subprocess.check_output(['xdrparser', 'XDRs/bucket-0001541ef1d33d6b76f70c2d5b4347e7e03a3c50faf7998689a179c369d88c97.xdr'])

print(len(output_1))
print(len(output_2))


```

```python
data_obj = json.loads(output_1)
display_json = JSON(data_obj)
# display(data_obj)
display(data_obj[0]['liveEntry']['data'])
# display(data_obj['liveEntry']['data'])
# dir(display_json)
# display(display_json)
```
